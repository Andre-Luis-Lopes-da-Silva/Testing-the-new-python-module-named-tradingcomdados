```python
from tradingcomdados import funds_data as fd
import pandas as pd
```

```python
fd.get_fund_data('2023-01-01', '2023-03-31', verbose = False)
# verbose = True --> enabled shows extraction progress
```

```python
# Get the information about the fund
alaska = fd.get_fund_data('2022-01-01', '2024-03-14', cnpj = '26.673.556/0001-32', verbose = True)
```

```python
# Plotting the quote variation
alaska.index = pd.to_datetime(alaska.DT_COMPTC)
```

```python
alaska.VL_QUOTA.plot()
```

```python
# Plotting the variation in the number of shareholders
alaska.NR_COTST.plot()
```

```python
# Variation of net worth
#alaska.VL_PATRIM_LIQ.plot();  # ; shows more information
alaska.VL_PATRIM_LIQ.plot();
```

```python
from tradingcomdados import alternative_data as ad
```

```python
# Obtaining composition of IBOV
ad.index_composition('ibov')
```

```python
# Obtaining composition of IFIX
ad.index_composition('ifix')
```

```python
ifix = ad.index_composition('ifix')
```

```python
# Getting the TOP 10 funds with largest contribution to the index
top10 = ifix.sort_values(by='part', ascending=False).head(10)
top10
```

```python
top10_ = ifix.sort_values(by='part', ascending=False)['part'].head(10)
top10_
```

```python
top10_c = ifix.sort_values(by='part', ascending=False)[['part', 'cod']].head(10)
top10_c
```
